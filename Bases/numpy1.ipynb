```python
import numpy as np
```

```python
# CREATION des tableau
a = np.array([1, 2, 3])
b = np.array([[1, 2, 3], [4, 5, 6]])
print("a:", a)
print("b:\n", b)
print("Type:", type(a))
print("ndim:", b.ndim, "shape:", b.shape, "size:", b.size) # Affichage de la dimension d'une table et de la  forme
print("dtype:", a.dtype)
```

```python
zeros = np.zeros((3, 3))
ones = np.ones((2, 4), dtype=int) # matrice constitue des un seulement
eye = np.eye(4)  # matrice identit√©
ar = np.arange(0, 10, 2)
lin = np.linspace(0, 1, 5)
rand = np.random.rand(3, 2) # generation des valeurs aleatoires
print("zeros:\n", zeros)
print("ones:\n", ones)
print("eye:\n", eye)
print("arange:", ar)
print("linspace:", lin)
print("rand:\n", rand)
```

```python
# INDEXATION, SLICING, MASQUES
# generations des valeurs allant de 0 a 19 et les passer dansun tableau numpy
arr = np.arange(20).reshape(4,5)
print("arr:\n", arr)
print("arr[1,2]:", arr[1,2])
print("arr[1]:", arr[1]) #Accessibilite a une valeur a partir de son index

```

```python
#Accessibilite a des valeur du tableau avec le slicing
print("arr[:,2]:", arr[:,2])
print("arr[1:3, 2:5]:\n", arr[1:3, 2:5])
```

```python
mask = arr % 2 == 0
# returne vrai si celui ci est pair ou faux dans le cas contraire
print("mask (pairs):\n", mask)
print("valeurs paires:", arr[mask]) #aficichage des reu=sultat des valeurs paires
```

```python
# Operation Arithmetique et trigonometrique
x = np.array([1,2,3,4])
y = np.array([5,6,7,8])
print(f"{x}  +  {y} =   ", x+y) # calcul de la somme
print(f"{x}  *  {y} = ", x*y)
print(f"{y}  /  {x} =  ", y/x)
print(f"{x}  **2   =  ", x**2)
print(f"sqrt({x}): ", np.sqrt(x)) # caclu de la racine carre
print(f"exp({x}):  ", np.exp(x))   # calcul de l'exponentielle de x
print(f"sin({x}):  ", np.sin(x)) # calcul du sin
```

```python
# Algebre lineaire
A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])
print("A:\n", A)
print("B:\n", B)
```

```python
print("Produit matriciel(A,B):\n", A.dot(B))  # Calcule matriciel
print("matmul(A,B):\n", A @ B)   # calcul du produit de 2 matrices
```

```python
print("detrminant(A):", np.linalg.det(A)) # calcul du determinant de la matrice
print("La matrice inverse de (A):\n", np.linalg.inv(A))  # calcul de la matrice inverse
```

```python
# Generation des valeurs aleatoires & statistiques
data = np.random.randint(0,100,size=(5,4))
print("data:\n", data)
print("mean:", data.mean()) # calcul de la moyenne
print("mean axis=0 :", data.mean(axis=0))  # calcul de la moyenne par colonne
print("sum axis = 1 (lignes):", data.sum(axis=1))  # calcul de la moyenne par ligne
print("median:", np.median(data)) # calcul du medianne
print("std:", np.std(data))
print("argmax:", data.argmax(), "argmin:", data.argmin()) # Affichage de la valeur maximale et minimal
```
